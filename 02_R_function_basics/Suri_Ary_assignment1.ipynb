### Data Set 1: NFL 2008 Field Goal Attempt Data

### Question 1: Which NFL Kicker made the highest percentage of their field goal attempts in the 2008 Season?

### Psuedocode:
- Group by Kicker Name
- Count total kicks made
- Count total kicks attempted
- Calculate conversion rate
- Add conversion rate to new table 
- Arrange in descending order based on conversion rate


```python
#read in file and print info 

nfl_fga = pd.read_csv("data/nfl2008_fga.csv")
print(nfl_fga.info())
```

```python
#Group data set based on kicker name
x1 = nfl_fga.groupby(['name'])
```

```python
#Get total made, total attempted, and assign to temp dataframe grouped by kicker name

temp_df = pd.DataFrame()
temp_df = temp_df.assign(made = x1.GOOD.sum(), 
                         total_attempted = x1['GOOD'].count()).assign(conversion_rate = lambda x: (x['made'] / x['total_attempted']))
temp_df.head()
```

```python
#Sort table in descending order by percent made
temp_df = temp_df.sort_values(by="conversion_rate", ascending=False)
temp_df.head()
```

```python
#Pipe together in function
def kicker_rates(df):
    x1 = df.groupby(['name'])
    temp_df = pd.DataFrame()
    temp_df = temp_df.assign(made = x1.GOOD.sum(), 
                         total_attempted = x1['GOOD'].count()).assign(conversion_rate = lambda x: (x['made'] / x['total_attempted'])).sort_values(by="conversion_rate", ascending=False)
    return temp_df 
   
print(kicker_rates(nfl_fga).head())
    
```

```python
#Sort by highest salary
salary_by_position = salary_by_position.sort_values("avg_salary", ascending=False)
salary_by_position
```

### Question 2: Is there a difference in free throw percentage based on positions in the NBA?

### Psuedocode: 
- Load dataset

- Calculate average free throw percentage by position

```python
#Group by position
position_NBA = nba.groupby("POSITION")
```

```python
#Calculate average free throw percentageand assign to new dataframe

ft_by_position = pd.DataFrame()
ft_by_position = ft_by_position.assign(num_players = position_NBA['POSITION'].count(), 
                                               avg_ft = position_NBA['FT_PC'].sum()/position_NBA['POSITION'].count())

ft_by_position
```

```python
#Sort by free throw percentage

ft_by_position = ft_by_position.sort_values("avg_ft", ascending=False)
ft_by_position
```

```python
#Pipe together in function
def ft_percentage_by_position(df):
    position_NBA = nba.groupby("POSITION")
    ft_by_position = pd.DataFrame()
    ft_by_position = ft_by_position.assign(num_players = position_NBA['POSITION'].count(), 
                                               avg_ft = position_NBA['FT_PC'].sum()/position_NBA['POSITION'].count())
    ft_by_position = ft_by_position.sort_values("avg_ft", ascending=False)
    return ft_by_position

ft_percentage_by_position(nba)
```
